1. What is the difference between a function and a method in Python?

Function: A function is a block of code that performs a specific task and is defined independently. It is called by its name, and it can take arguments as input.

Example: def add(x, y): return x + y

Method: A method is a function that is associated with an object (an instance of a class). It is called on an object using dot notation (.).

Example: my_list.append(5) (append is a method of the list object)

Key Difference: A function is a standalone block, while a method operates on an object and has access to the object's internal data.

5. What are iterators in Python, and how do they differ from iterables?

* **Answer:**
    * **Iterables:** An iterable is any object that can be looped over (e.g., lists, tuples, strings, dictionaries, sets). It has the ability to return an iterator.
        * **Example:** `my_list = [1, 2, 3]` is an iterable
    * **Iterators:** An iterator is an object that provides a way to access elements of an iterable sequentially. It has two methods: `__iter__()` (which returns the iterator object itself) and `__next__()` (which returns the next element, raising `StopIteration` when no more elements).
       * **Example:** `my_iterator = iter(my_list)` is an iterator created from an iterable list, using `__iter__` method internally
  * **Difference:** Iterables are data containers; iterators provide an interface to sequentially go through the iterable's data. An iterable *can create* an iterator.

6. Explain the concept of generators in Python and how they are defined.

Answer:

Generators: Generators are a special type of function that use the yield keyword instead of return.

Lazy Evaluation: Generators don't compute all values at once; they produce values on demand (lazily) each time they are iterated over. This makes them very memory-efficient for large sequences.

Definition: A generator function is defined like a regular function but uses the yield keyword to produce a value. When a generator function is called, it returns a generator object (an iterator).

Exampledef my_generator(n):
   for i in range(n):
      yield i * 2

gen = my_generator(5)
for item in gen:
   print(item)

7. What are the advantages of using generators over regular functions?

Answer:

Memory Efficiency: Generators use lazy evaluation which is a memory-saving technique for big datasets since it doesn't store all values in memory; they yield one by one.

Better Performance: In cases with very large datasets, generators are much more performant.

Readability for Sequence Generation: For generating sequences of values (e.g. Fibonacci sequence) generator code can be more readable.

No need for a separate collection: For iteration over large datasets, a generator will generate elements of that dataset without storing a collection in memory and thus can save memory.

10. What is the difference between map(), reduce(), and filter() functions in Python?

Answer:

map():

Applies a function to each element of an iterable and returns a map object (iterator).

Output size is the same as the input iterable, but values may change after transformation

reduce():

Applies a function cumulatively to the items of an iterable, reducing the iterable to a single value. It's located in the functools library.

Output is a single value.

Requires import: from functools import reduce

filter():

Filters elements of an iterable based on a condition (a function that returns True or False) and returns a filter object (iterator) with the elements that satisfy the condition.

Output is a smaller or equal sized iterable with the filtered elements

# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list

````python
def sum_of_even(numbers):
    return sum(num for num in numbers if num % 2 == 0)
print(sum_of_even([1,2,3,4,5,6]))
````

# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

````python
def is_prime(num):
    if num <= 1 or num > 200:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True
print(is_prime(35))
print(is_prime(137))
````

# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms

````python
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.current = 0
        self.next_val = 1
        self.count = 0
    def __iter__(self):
        return self
    def __next__(self):
        if self.count < self.n:
            result = self.current
            self.current, self.next_val = self.next_val, self.current + self.next_val
            self.count += 1
            return result
        else:
            raise StopIteration

fib_iter = FibonacciIterator(10)
for num in fib_iter:
    print(num)

# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

for power in power_of_2(5):
    print(power)

# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def file_reader(file_name):
   with open(file_name, "r") as file:
       for line in file:
           yield line.strip() # Remove trailing newline character

# Example Usage
# Create a test file named "test_file.txt" with several lines
# with open("test_file.txt", "w") as f:
#     f.write("This is line 1.\n")
#     f.write("This is line 2.\n")
#     f.write("This is line 3.\n")
for line in file_reader("test_file.txt"):
    print(line)
````

# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

````python
def remove_vowels(input_str):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda char: char not in vowels, input_str))
print(remove_vowels("hello world"))
````

# 11. Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,-  if the value of the order is smaller than 100,00 . Write a Python program using lambda and map.

````python
data = [
    ["34587", "Learning Python, Mark Lutz", 4, 40.95],
    ["98762", "Programming Python, Mark Lutz", 5, 56.80],
    ["77226", "Head First Python, Paul Barry", 3, 32.95],
    ["88112", "Einfhrung in Python3, Bernd Klein", 3, 24.99],
]

def process_order(item):
  order_number, _, quantity, price = item
  product = quantity * price
  if product < 100:
      product += 10
  return (order_number, product)

result_list = list(map(process_order, data))
print(result_list)
````
